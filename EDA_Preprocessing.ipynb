```python
# Import Pandas
import pandas as pd
#Import Numpy for numerical computation
import numpy as np
# Import matplotlib & seaborn for visualisation
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as st
import missingno as msno
```

```python
#Import Dataset
dataset = pd.read_csv('X1_Log.csv')
dataset
```

```python
#Dataset Main Statistics
dataset.describe()
```

```python
#Replace -999 with np.NaN for computation
dataset = dataset.replace([-999, 0], np.NaN)
dataset
```

```python
#Table of Missing Value
total = dataset.isnull().sum().sort_values(ascending=False)
percent = (dataset.isnull().sum()/dataset.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1,join='outer', keys=['Total Missing Count', '% of Total Observations'])
missing_data.index.name =' dataset'

missing_data.head(20)
```

```python
#Bar Chart of Missing Value
missing_values = dataset.isnull().sum(axis=0).reset_index()
missing_values.columns = ['column_name', 'missing_count']
missing_values = missing_values.loc[missing_values['missing_count']>0]
missing_values = missing_values.sort_values(by='missing_count')

ind = np.arange(missing_values.shape[0])
width = 0.1
fig, ax = plt.subplots(figsize=(12,3))
rects = ax.barh(ind, missing_values.missing_count.values, color='b')
ax.set_yticks(ind)
ax.set_yticklabels(missing_values.column_name.values, rotation='horizontal')
ax.set_xlabel("Missing Observations Count")
ax.set_title("Missing Observations Count - Numeric Features")
plt.show()
```

```python
#Matrix of Missing Value
msno.matrix(dataset)
```

```python



#Imputing Missing Values
#Median Values Were Used To Impute Missing Values For Features Which Has A Small Size of Missing Values
#Compared To The Overall Dataset Size, Thus Keeping Bias At A Minimal


```

```python
#Replace Missing Value with Median Value
median_value=dataset['LL9D'].median()
dataset['LL9D']=dataset['LL9D'].fillna(median_value)
```

```python
#Replace Missing Value with Median Value
median_value=dataset['LL9S'].median()
dataset['LL9S']=dataset['LL9S'].fillna(median_value)
```

```python
#Replace Missing Value with Median Value
median_value=dataset['MSFL'].median()
dataset['MSFL']=dataset['MSFL'].fillna(median_value)
```

```python
#Replace Missing Value with Median Value
median_value=dataset['GrC'].median()
dataset['GrC']=dataset['GrC'].fillna(median_value)
```

```python
#Replace Missing Value with Median Value
median_value=dataset['DENS'].median()
dataset['DENS']=dataset['DENS'].fillna(median_value)
```

```python
#Replace Missing Value with Forward Fill, FFill Was Used Since SW Instances Have 
#Visible Correlations With Their Preceding Values
dataset['SW']=dataset['SW'].fillna(method='ffill')
```

```python
#Correlation Table of Features
correlation = dataset.corr()
correlation
```

```python



#Creating Dataset For Computing Missing Values For VCL since VCL has a lot of missing values
#which cannot be imputed with median values which will introduce disproportionate bias


```

```python
#Correlation with Respect to VCL
print(correlation['VCL'].sort_values(ascending = False),'\n')
```

```python
#VCL Dataset - Only With Respect To The Highest Correlated Feature ie GrC
dataset2 = dataset.drop(['DENS', 'SONI', 'SW', 'MSFL', 'CNL_dec', 'LL9S', 'DEPTH', 'LL9D'], axis=1)
dataset2
```

```python
#VCL Dataset - Only With Non Null Values
numeric_features_lnrreg = dataset2[pd.notnull(dataset2['VCL'])]
numeric_features_lnrreg
```

```python
#Reshaping Dataset
X = numeric_features_lnrreg['GrC'].values.reshape(-1,1)
y = numeric_features_lnrreg['VCL'].values
```

```python



#Training Linear Regression Model to Compute Missing VCL Data
#Linear Regression was chosen since GrC has a high linear correlation with VCL


```

```python
#Importing Linear Regression Model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Fitting Training Data
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#Predict Test Data
y_pred = lr.predict(X_test)
```

```python
#Evaluate Performance
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#Convert To Panda Series
VCL_pred = pd.Series(y_pred2.astype(float).tolist())
VCL_pred
```
